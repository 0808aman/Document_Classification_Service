```python
from bs4 import BeautifulSoup
import os
import glob
import pandas as pd
from pathlib import Path
import re
import requests
```

```python
def text_extract(file):
  soup = BeautifulSoup(file, 'html.parser')
  
  page = soup.find('div', class_ = "ocr_page")
  if page != None:
    text = page.get_text()
  else:
    return None
  return text
```

```python
# file_list = glob.glob(folder_path + "/*.hocr")
# file_list[0].split('/')[-1].split('.')[0]
```

```python
folder_path = '/content/drive/MyDrive/NPL/HOCR Files'
file_list = glob.glob(folder_path + "/*.hocr")
dict_ = {'Name': [], 'Text': [], 'Class': []}
  
for i in range(0,len(file_list)):
  with open (file_list[i], 'r') as fr:
    content = fr.read().strip()
    text = text_extract(content)
    if text != None:
      dict_["Text"].append(text)
      dict_["Name"].append(file_list[i].split('/')[-1].split('.')[0])
      dict_["Class"].append(file_list[i].split('/')[-1].split('.')[0])

```

```python
data = pd.DataFrame(dict_)
```

```python
data.head()
```

```python
# def forma(text):
#   if text.startswith('bc'):
#     return "Birth_Certificate"
#   elif text.startswith('mc'):
#     return "Marrige_Certificate"
#   elif text.startswith('ft'):
#     return "Tax Form"


def forma(text):
    d = {"bc": "Birth Certificate", "mc": "Marriage Certificate","ft": "FTR"}
    return d[text[:2]]
```

```python
data['m_class'] = data['Class'].apply(lambda x: forma(x))
```

```python
import re
def text_cleaning(text):
    # text = text.lower()
    # text = re.sub('},', '\n', text)
    # text = re.sub('{', '', text)
    # text = re.sub('}}', '', text)
    text = re.sub(':', ',', text)
    words = text.split("\n")
    #words = [word for word in words if (not word.startswith("https")) and (not word.startswith("@"))]
    words = [word for word in words if len(word) > 2]
    text = " ".join(words)
    text = re.sub(r"[,.;:-_+={}[]()%@#?!&$/]+\ *", " ", text)
    text = re.sub(r'[^\x00-\x7F]+'," ", text)
    #text = text.replace("'", " ").replace("\r", " ").replace("\n", " ").replace("/", " ")
    text = re.sub(' +', " ", text)
    text = " ".join([word for word in text.split(" ") if len(word) > 2])

    return text
```

```python
data_new['clean_text'] = data_new['Text'].apply(lambda x: text_cleaning(x))
data_new.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
A = vectorizer.fit_transform(data_new["clean_text"])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
b = le.fit_transform(data_new["m_class"].values)
```

```python
from sklearn.model_selection import train_test_split
A_train, A_test, b_train, b_test = train_test_split( A,b, test_size=0.2, random_state=0)
```
